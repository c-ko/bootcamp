```python
import numpy as np
import pandas as pd

df = pd.read_csv('data/gfmt_sleep.csv',na_values = '=')
df['insomnia']=df['sci']<= 16
df.head()
```

# Split Apply Combine 

```python
grouped = df.groupby('insomnia')

grouped 
```

```python
df_median = grouped.median().reset_index()
```

```python
df_median
```

```python
df_gender_insom_med = df.groupby(['gender','insomnia']).median().reset_index()
```

```python
df_gender_insom_med
```

```python
grouped = df.groupby('insomnia')

grouped['percent correct'].rank(method='first') #person's index on the left; person's rank on the right; .5 is because they tied 
```

```python
df.head()
```

```python
grouped.rank?
```

```python
def coeff_of_var(data):
    """Compute cpefficient of variation from an array of data"""
    return np.std(data) / np.mean(data)
```

```python
grouped.agg(coeff_of_var).reset_index() #agg means aggregate 
```

```python
grouped['percent correct'].transform(np.sin)
```
