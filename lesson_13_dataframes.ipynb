```python
import numpy as np
import pandas as pd
```

```python
!head data/gfmt_sleep.csv 

df = pd.read_csv('data/gfmt_sleep.csv', na_values='*')

type(df)
```

```python
df.head()
```

```python
# Index by column
df['percent correct'][4]
```

```python
# Index by row index followed by column index
#.loc means index by row and then by column
df.loc[4, 'percent correct']
```

```python
df['participant number'] == 8
```

```python
df.loc[df['participant number'] == 8, 'percent correct']
```

```python
df.loc[(df['participant number'] == 8), :]
```

```python
# All women under 21

df.loc[(df['age']<21) & (df['gender'] == 'f'), :]
```

```python
inds = (df['age']<30) & (df['gender'] == 'f') & (df['percent correct']>99)
df.loc[inds ,:]
```

```python
df['insomnia'] = df['sci'] <= 16 #make a new column in the dataframe and population it whenver sci <= 16
```

```python
np.mean(df.loc[df['insomnia'], 'percent correct'])
```

```python
np.mean(df.loc[~df['insomnia'], 'percent correct'])
```

```python
df.describe()
```

```python
df.to_csv("gfmt_sleep_with_insomnia.csv", index=False)
```
